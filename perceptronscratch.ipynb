```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("placement (2).csv")
```

```python
epoch = 1000
```

```python
x = df.iloc[:, :2]
y = df.iloc[:, -1]
```

```python
x = np.array(x)
y = np.array(y)
```

```python
import random
def step(z):
  if( z > 0):
    return 1
  else:
    return 0

def Perceptron(x,y):
  x = np.insert(x, 0, 1, axis=1)
  weights = np.ones(x.shape[1])
  lr = 0.1
  for i in range(epoch):
    choice = random.randint(0,len(x) - 1)
    output = step(np.dot(weights, x[choice]))
    if(output != y[choice] and output < y[choice]):
      weights = weights + lr*(x[choice])
    elif(output == y[choice]):
      pass
    else:
      weights = weights - lr*(x[choice])

  return weights[0], weights[1:]

```
